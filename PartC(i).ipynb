```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load data from the CSV file
df = pd.read_csv('key_comparisons.csv')

# Extract sizes and key comparisons
sizes = df['Size']
comparisons = df['KeyComparisons']

# Plot empirical results
plt.figure(figsize=(10, 6))
plt.plot(sizes, comparisons, 'o-', label='Empirical Results', color='blue')
plt.xlabel('Size of the array (n)')
plt.ylabel('Number of key comparisons')
plt.title('Empirical Results of Hybrid Algorithm')
plt.grid(True)
plt.legend()

# Set y-axis limits (optional, but ensures both plots have the same range)
plt.ylim(min(comparisons.min(), theoretical_time.min()), max(comparisons.max(), theoretical_time.max()))

plt.show()
```
