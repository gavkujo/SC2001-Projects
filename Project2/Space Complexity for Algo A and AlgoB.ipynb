```python
import numpy as np
import matplotlib.pyplot as plt
import heapq  # Import heapq for the min-heap functionality

# Function to test space complexity of Algorithm A and B
def test_space_complexity():
    vertices = [100, 200, 500, 1000, 2000]  # List of vertices to test on
    edges = [200, 400, 1000, 2000, 4000]  # Example list of edges corresponding to each vertex count
    space_algo_a = []  # Space for Algorithm A
    space_algo_b = []  # Space for Algorithm B
    
    for i, V in enumerate(vertices):
        E = edges[i]  # Get the number of edges for the current graph size
        
        # Space for Algorithm A (2 arrays: dist and visited)
        space_a = V * 2
        
        # Space for Algorithm B (dist array + min-heap + adjacency list)
        # Assuming adjacency list, space complexity is O(V + E)
        space_b = V * 2 + E  # Space for Algorithm B (dist + min-heap + adjacency list)
        
        # Append space usage to lists
        space_algo_a.append(space_a)
        space_algo_b.append(space_b)
        
        print(f"Vertices: {V}, Edges: {E}, Algorithm A Space: {space_a} units, Algorithm B Space: {space_b} units")

    return vertices, space_algo_a, space_algo_b

# Plotting memory usage comparison
def plot_space_complexity(vertices, space_algo_a, space_algo_b):
    plt.plot(vertices, space_algo_a, label='Algorithm A (Array-based)', marker='o')
    plt.plot(vertices, space_algo_b, label='Algorithm B (Min-Heap-based)', marker='x')

    plt.xlabel('Number of Vertices (V)')
    plt.ylabel('Memory Usage (units)')
    plt.title('Memory Usage Comparison: Algorithm A vs Algorithm B')
    plt.legend()
    plt.grid(True)
    plt.show()

# Run the test and plot the results
vertices, space_algo_a, space_algo_b = test_space_complexity()
plot_space_complexity(vertices, space_algo_a, space_algo_b)
```
