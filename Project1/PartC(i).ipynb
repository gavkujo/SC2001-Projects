```python
# Read the CSV file
data = pd.read_csv("optimal_S.csv")

# Plot the data for each input size
plt.figure(figsize=(10, 6))

# Get the unique input sizes
input_sizes = data['InputSize'].unique()

# Plot for each input size
for input_size in input_sizes:
    subset = data[data['InputSize'] == input_size]
    plt.plot(subset['S'], subset['KeyComparisons'], label=f"Input Size: {input_size}")

# Add labels and title
plt.xlabel("Threshold S")
plt.ylabel("Key Comparisons")
plt.title("Key Comparisons vs Threshold S for Different Input Sizes")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
