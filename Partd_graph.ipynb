```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv("Partd_key_comparisons.csv")
```

```python
# Plot settings
plt.figure(figsize=(14, 6))

# Graph 1: Number of key comparisons vs. array size
plt.subplot(1, 2, 1)
plt.plot(data['Size'], data['HybridKeyComparisons'], label='Hybrid Merge Sort', color='blue')
plt.plot(data['Size'], data['MergeSortKeyComparisons'], label='Standard Merge Sort', color='green')
plt.xlabel('Array Size')
plt.ylabel('Number of Key Comparisons')
plt.title('Key Comparisons vs. Array Size')
plt.legend()

# Graph 2: CPU Time vs. array size
plt.subplot(1, 2, 2)
plt.plot(data['Size'], data['HybridTime'], label='Hybrid Merge Sort', color='blue')
plt.plot(data['Size'], data['MergeSortTime'], label='Standard Merge Sort', color='green')
plt.xlabel('Array Size')
plt.ylabel('CPU Time (seconds)')
plt.title('CPU Time vs. Array Size')
plt.legend()

# Display the plots
plt.tight_layout()
plt.show()
```
